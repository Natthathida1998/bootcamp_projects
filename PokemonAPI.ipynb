# **Pokemon** **API** **Project**

```python
import requests 
import time
import pandas as pd
```

```python
ids = []
names = []
heights = []
weights = []

for i in range(1,21):
    url = f"https://pokeapi.co/api/v2/pokemon/{i}/"
    resp = requests.get(url)
    result = resp.json()
    ids.append(result['id'])
    names.append(result['name'])
    heights.append(result['height'])
    weights.append(result['weight'])
    time.sleep(1) #second to break

df = pd.DataFrame({
    "id": ids,"name": names, "height": heights, "weight": weights,
})

df
```
